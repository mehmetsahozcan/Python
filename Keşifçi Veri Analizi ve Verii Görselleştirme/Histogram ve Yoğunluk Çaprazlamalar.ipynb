```python
sns.kdeplot(df.yaş, shade = True);
```

```python
(sns
 .FacetGrid(df,
             hue = "göğüsağrısı",
             height = 10,
             xlim  = (0,500),
             )
 .map(sns.kdeplot, "kanbasıncı", shade = True )
 .add_legend()
);
```

Göğüs ağrısı tiplerine göre kan basıncı değerlerine baktığımızda görüyoruz ki;
ATYPICAL ANGINA diğerlerine göre çok daha fazla yer kaplıyor.
Sonrasında ise NON-ANGİNAL PAIN,  en az olan ise Asymptomatıc


Bunu değerlerini göstererek kanıtlayalım

```python
df.groupby(["göğüsağrısı"])["kanbasıncı"].mean()
```

```python
 sns.catplot( x = "göğüsağrısı",  y = "kalpatıs", hue = "cinsiyet", kind = "point", data  = df);
```
