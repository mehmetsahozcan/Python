### Çaprazlamalar

```python
sns.catplot(x = "göğüsağrısı", y = "yaş", data = df)
```

```python
# Asymptomatic tipi göğüs ağrısına baktığımızda 50-68 yaş arasında yoğunluk göstermekte.
#Atypıcal angina tipi göğüs ağrısı 55-60  yaş arasında yoğu nluk göstermektedir
# Non anginal pain tipine baktığımızda ise 45-52 arasında olduğunu söyleyebiliriz.
```

```python
sns.catplot(x = "cinsiyet", y = "kalpatıs", data = df)
```

```python
#Erkeklerde kalp atışı yoğunluğu 140-180 arasında iken kadınlarda bu değer 160-175 arasındadır.
```

```python
sns.barplot(x = "göğüsağrısı" , y = "yaş" , hue = "cinsiyet", data = df);  #boyut eklemek için hue kullanıldı
```

```python
sns.barplot(x = "göğüsağrısı", y = "kanbasıncı", hue = "cinsiyet", data = df  )
```

```python
sns.barplot(x = "cinsiyet", y = "kalpatıs", hue = "göğüsağrısı" , data = df);
```

```python
df.groupby(["cinsiyet","göğüsağrısı"])["kalpatıs"].mean() #gruplayarak doğruladık
```

## Histogram ve Yoğunkluk Grafikleri

Sayısal değişkenleri  görselleştirmek için kullanılır. Dağılım hakkında bilgi verir

```python
sns.distplot(df.yaş, kde =  False ,  color = "g"); #kde nin false yapmamız yoğunluk çizgisini istemediğimizden
```

```python
sns.distplot(df.yaş, color = "g" , hist= False);  #Sadece yoğunluk grafiği için hist = false deriz.
```

```python
sns.kdeplot(df.yaş, color = "b" ,shade = True);
```
